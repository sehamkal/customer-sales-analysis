```python


import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

df.head()

df.groupby("day")["total_bill"].sum()
sales_by_day = df.groupby("day")["total_bill"].sum()
sales_by_day.plot(kind="bar")
plt.title("Total Revenue by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")

plt.show()

df.groupby(["day", "sex"])["total_bill"].sum()

data = df.groupby(["day", "sex"])["total_bill"].sum().unstack()
data.plot(kind="bar")
plt.title("Revenue by Day and Gender")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```
